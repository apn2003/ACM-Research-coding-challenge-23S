```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os

def arrFunc(arr):
    equal = 0
    dic = {'O' : 0, 'B' : 0, 'A' : 0, 'F' : 0, 'G' : 0, 'K' : 0, 'M' : 0}
    for i in arr:
        if(i == 'O'):
            dic['O'] += 1
        elif(i == 'B'):
            dic['B'] += 1
        elif(i == 'A'):
            dic['A'] += 1
        elif(i == 'F'):
            dic['F'] = 1
        elif(i == 'G'):
            dic['G'] += 1
        elif(i == 'K'):
            dic['K'] += 1
        else:
            dic['M'] += 1

    result = 'O'
    big = dic['O']
    for i in dic:
        if(dic[i] > big):
            big = dic[i]
        elif(dic[i] == big):
            equal += 1
       
    if(equal == 7):
        result = "All the spectral classes are equally likely to be a part of "
    else:
        result = ""
        for i in dic:
            if(dic[i] == big):
                result += (i + ",")
        result = " will be mostly likely classified as the spectral class: " + result
    
    return result
  
data = pd.read_csv("/kaggle/input/star-dataset/6 class csv.csv")

starDict = {"Brown Dwarf" : 0, "Red Dwarf" : 0, "White Dwarf" : 0, "Main Sequence" : 0, "Supergiant" : 0, "Hypergiant" : 0}

for number in data['Star type']:
    if(number == 0):
        starDict["Brown Dwarf"] += 1;
    elif(number == 1):
        starDict["Red Dwarf"] += 1;
    elif(number == 2):
        starDict["White Dwarf"] += 1;
    elif(number == 3):
        starDict["Main Sequence"] += 1;
    elif(number == 4):
        starDict["Supergiant"] += 1;
    else:
        starDict["Hypergiant"] += 1;

big = starDict["Brown Dwarf"]
equal = 0
star = "Brown Dwarf"

for i in starDict:
    if(starDict[i] > big):
        star = i
        big = starDict[i]
    elif(starDict[i] == big):
        equal = equal + 1;
        
if(equal == 6):
    print("There is an equal amount of stars in each category\n")
elif(equal > 1 and equal < 6):
    star = ""
    for i in starDict:
        if(starDict[i] == big):
            star += (i + ", ")
    print(star + "are the categories with the largest amount of stars "  + star + " have " + big + " stars each.\n")
else:
    print(star + " category has the largest amount of stars. " + star + " has " + big + " stars.\n")

arr0, arr1, arr2, arr3, arr4, arr5 = [], [], [], [], [], []
i = 0

for number in data['Star type']:
    if(number == 0):
        arr0.append(data['Spectral Class'][i])
    elif(number == 1):
        arr1.append(data['Spectral Class'][i])
    elif(number == 2):
        arr2.append(data['Spectral Class'][i])
    elif(number == 3):
        arr3.append(data['Spectral Class'][i])
    elif(number == 4):
        arr4.append(data['Spectral Class'][i])
    else:
        arr5.append(data['Spectral Class'][i])
    
    i += 1

print("Brown Dwarf" + arrFunc(arr0))
print("Red Dwarf" + arrFunc(arr1))
print("White Dwarf" + arrFunc(arr2))
print("Main Sequence" + arrFunc(arr3))
print("Supergiant" + arrFunc(arr4))
print("Hypergiant" + arrFunc(arr5))
```
